# Preperation and Overview

### Buisiness Use

Since the COVID-19 pandemic, working remotely has exploded in popularity in the workforce. What started off as a forced pivot due to a gloabl pandemic has now turned into the perferred work style for thousands of employees. Identifying the long term effects of working remotely is very important for companies who are experiencing the shift towards working from home. This dataset holds information about 150 remote workers, tracked over 90 days. The attributes are sepereated into two categories: Behavioral and Psychological. Behavioral attributes were gathered by a background python processes, and the psychological gathered via daily surveys and the UCLA loneliness scale. Each attribute is used to find three classes, burnout risk, which can be low, medium and high. The end goal of this dataset is to be able to predict burnout ahead of time, so that it can be subdued before it occurs. The results can be very useful for companies that have remote employees to improve retention, prevent burnout, and prevent productivity loss.

How well the algorithm needs to perform: According to the dataset source, each employee burnout costs  7,000− 13,000 dollars due to the loss of productivity. The source also states that the algorithim is expected to have 80-86% accuracy. Lets say a company has 150 employees, with 40 of them medium-high burnout risk. At 80% accuracy, the algorithim would be able to predict 32 employees that are medium-high risk. Based on the burnout costs, the company could potentially save  224,000− 442,000 dollars by intervening early. This means that the algorithim does not need to be 100% accurate to be useful, 80% is enough to make a positive impact.

##Binary Logistic Regression

```python
import numpy as np
from scipy.optimize import minimize
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from scipy.special import expit
from numpy.linalg import pinv


class CustomBinaryLogisticRegressionBase:
    def __init__(self, eta, solver='steepest_ascent', max_iter=20, C=1.0, tol=1e-4, mini_batch_size=32, regularization='none'):
        self.eta = eta
        self.max_iter = max_iter
        self.C = C
        self.solver = solver
        self.tol = tol
        self.mini_batch_size = mini_batch_size
        self.regularization = regularization
        self.classifiers_ = []

    def __str__(self):
        return 'Base Binary Logistic Regression Object, Not Trainable'

    # convenience, private and static:
    @staticmethod
    def _sigmoid(theta):
        # increase stability, redefine sigmoid operation
        return expit(theta) #1/(1+np.exp(-theta))

    @staticmethod
    def _add_intercept(X):
        return np.hstack((np.ones((X.shape[0],1)),X)) # add bias term

    # public:
    def predict_proba(self, X, add_intercept=True):
        # add bias term if requested
        Xb = self._add_intercept(X) if add_intercept else X
        return self._sigmoid(Xb @ self.weights) # return the probability y=1

    def predict(self, X):
        return (self.predict_proba(X) > 0.5) #return the actual prediction

    def gradient(self, X, y):
        # vectorized gradient: (1/m) * X.T @ (y - h)
        m = len(y)
        h = self._sigmoid(X @ self.weights)
        grad = (1 / m) * (X.T @ (y - h))

        if self.regularization == 'l1':
            reg = (self.C / m) * np.sign(self.weights)
            reg[0] = 0  # do not regularize bias
            grad -= reg
        elif self.regularization == 'l2':
            reg = (self.C / m) * self.weights
            reg[0] = 0  # do not regularize bias
            grad -= reg
        elif self.regularization == 'both':
            reg = (self.C / m) * np.sign(self.weights)
            reg[0] = 0  # do not regularize bias
            grad -= reg
            reg = (self.C / m) * self.weights
            reg[0] = 0  # do not regularize bias
            grad -= reg
        else:
            reg = np.zeros(self.weights.shape)

        return grad

    def fit(self, X, y):
        Xb = self._add_intercept(X) # add bias term
        num_samples, num_features = Xb.shape
        self.weights = np.zeros(num_features, dtype=float)

        if self.solver == 'steepest_ascent':
            self.steepest_ascent(Xb, y)
        elif self.solver == 'newton':
            self.newton(Xb, y)
        elif self.solver == 'stochastic_gradient_ascent':
            self.stochastic_gradient_ascent(Xb, y)

    def steepest_ascent(self, X, y): # Optimize with steepest ascent
        for _ in range(self.max_iter):
            grad = self.gradient(X, y)
            self.weights += self.eta * grad
            if np.max(np.abs(grad)) < self.tol:
                break

    def stochastic_gradient_ascent(self, X, y):
        for _ in range(self.max_iter):
            idxs = np.random.choice(len(y), self.mini_batch_size)
            grad = self.gradient(X[idxs], y[idxs])
            self.weights += self.eta * grad
            if np.max(np.abs(grad)) < self.tol:
                break

    def newton(self, X, y):
        for _ in range(self.max_iter):
            g = self.predict_proba(X, add_intercept=False).ravel() # get sigmoid value for all classes
            hessian = X.T @ np.diag(g*(1-g)) @ X # calculate the hessian
            if self.regularization in ['l2', 'both']:
                hessian -= 2 * self.C

            grad = self.gradient(X, y)
            self.weights += (pinv(hessian) @ grad)
            if np.max(np.abs(grad)) < self.tol:
                break


blr = CustomBinaryLogisticRegressionBase(0.1)
print(blr)
```

## MultiClass Logistic Regression: One versus All

```python
class MultiClassLogisticRegression:
    def __init__(self, eta, solver='steepest_ascent', max_iter=20, C=1.0, tol=1e-4, mini_batch_size=32, regularization='none'):
        self.eta = eta
        self.max_iter = max_iter
        self.C = C
        self.solver = solver
        self.tol = tol
        self.mini_batch_size = mini_batch_size
        self.regularization = regularization
        self.classifiers_ = []

    def __str__(self):
        if(hasattr(self,'w_')):
            return 'MultiClass Logistic Regression Object with coefficients:\n'+ str(self.w_) # is we have trained the object
        else:
            return 'Untrained MultiClass Logistic Regression Object'

    @property
    def coef_(self):
        if(hasattr(self,'w_')):
            return self.w_[:,1:]
        else:
            return None

    @property
    def intercept_(self):
        if(hasattr(self,'w_')):
            return self.w_[:,0]
        else:
            return None

    def fit(self,X,y):
        num_samples, num_features = X.shape
        self.unique_ = np.unique(y) # get each unique class value
        num_unique_classes = len(self.unique_)
        self.classifiers_ = [] # will fill this array with binary classifiers

        for i,yval in enumerate(self.unique_): # for each unique value
            y_binary = (y==yval) # create a binary problem
            # train the binary classifier for this class
            blr = CustomBinaryLogisticRegressionBase(eta=self.eta, solver=self.solver,
                                          max_iter=self.max_iter, C=self.C,
                                          tol=self.tol, mini_batch_size=self.mini_batch_size,
                                          regularization=self.regularization)
            blr.fit(X,y_binary)
            # add the trained classifier to the list
            self.classifiers_.append(blr)

        # save all the weights into one matrix, separate column for each class
        self.w_ = np.hstack([x.weights for x in self.classifiers_]).T

    def predict_proba(self,X):
        probs = []
        for blr in self.classifiers_:
            probs.append(blr.predict_proba(X).reshape(-1,1)) # get probability for each classifier

        return np.hstack(probs) # make into single matrix

    def predict(self,X):
        return self.unique_[np.argmax(self.predict_proba(X),axis=1)] # take argmax along row

lr = MultiClassLogisticRegression(eta=0.1, max_iter=1500)
print(lr)
```

##Training
Now let's find the best optimization technique and regularization term for the dataset.

Let's start with an initial run

```python
lr = MultiClassLogisticRegression(eta=0.1, max_iter=500)
lr.fit(X_train, y_train)
yhat = lr.predict(X_test)
print('Accuracy:', accuracy_score(y_test, yhat))
```

Now let's test with different C and solver values

```python
import time

solvers = ['steepest_ascent', 'stochastic_gradient_ascent', 'newton']
C_vals = [0.001, 0.01, 0.1, 1.0, 10.0]

results = []

for solver in solvers:
    for C in C_vals:
      lr = MultiClassLogisticRegression(eta=0.1, max_iter=500, solver=solver, C=C, regularization='l2')
      start = time.time()
      lr.fit(X_train, y_train)
      train_time = time.time() - start
      yhat = lr.predict(X_test)
      accuracy = accuracy_score(y_test, yhat)
      results.append({'solver': solver, 'C': C, 'accuracy': accuracy, 'training_time': train_time})

      print(f'Solver: {solver}, C: {C}, Accuracy: {accuracy}, Time: {train_time:.4f}s')
```

```python
import plotly.graph_objects as go
import pandas as pd

results_df = pd.DataFrame(results)

fig = go.Figure()

for solver in solvers:
    solver_df = results_df[results_df['solver'] == solver]
    fig.add_trace(go.Scatter(x=solver_df['C'], y=solver_df['accuracy'], mode='lines+markers', name=solver))

fig.update_layout(title='Accuracy vs. C for Different Solvers on l2',
                  xaxis_title='C',
                  yaxis_title='Accuracy',
                  xaxis_type='log')

fig.show()

#Training time plot
fig2 = go.Figure()

for solver in solvers:
    solver_df = results_df[results_df['solver'] == solver]
    fig2.add_trace(go.Scatter(x=solver_df['C'], y=solver_df['training_time'], mode='lines+markers', name=solver))

fig2.update_layout(title='Training Time vs. C for Different Solvers on l2', xaxis_title='C', yaxis_title='Training Time (s)', xaxis_type='log')
fig2.show()

```

Looking at the performace of accuracy and training time between the three:

###Steepest ascent
Steepest ascent has our best accuracy of .89 with C = .01 and 1. It also generally seems to perform the best out of the three procedures, only falling behind in accuracy at the lowest and middle range with C at .001 or .01. The time taken to train was generally along the same length as Stochastic, only taking a bit longer with the highest C and therefore the least penalizing regulation.

###Stochastic gradient
Stochastic gradient notably suffers with particularly high or low Cs, sometimes ouright falling appart at the steeper ends. It doesn't seem to fare well with heavy or light regularization, and only surpasses other optimizations at a C of 0.1. It's training time matches up with Steepest ascent and is quite low thanks to mini batching speeding up the process.

###Newton
Newton faces the least alterations per C value, staying at around a .86 accuracy for each C. Regularization doesn't seem to have a large effect on it's output largely due to the use of the hessian function. It's hessian calculation is also determining it's very long training time, setting it far apart from the other two optimizations


###Conclusion
Since training times are relatively similiar between the lower two, and steepest ascent's avergae accuracy is more reliable, we will now compare different regularization techniques to find our best performance.

After comparing between using l1, l2, and both, it seems that l2 at a C of 1.0 is still the most accurate option. L1's results peak at an average regularization of 0.1 and seem to worsen off the higher C gets. Using both is most similiar to l2 but immediately starts to dovetail at higher values past 0.1 . Training time is negligble here but l2 is generally the fastest, only facing higher rates when C is at it's lowest.

As such, we have chosen Steepest Ascent with a C value of 1.0 using l2 regularization

##Note about Paramter usage (data snooping)

Our parameter selection does involve the evaluation of C and the solver on the test set itself. This can lead to an idealistic estimation of general performance and is called data snooping. We acknowledge that a more proper approach would be to cross validate the training set and ensure that all test data is unused until the model must be evaluated.

##Compare with scklearn
now let's compare this performance with sklearn

```python
import time
from sklearn.linear_model import LogisticRegression as SKLogisticRegression


start = time.time()
lr_best = MultiClassLogisticRegression(eta=0.1, max_iter=500, solver='steepest_ascent', C=1.0, regularization='l2')
lr_best.fit(X_train, y_train)
custom_time = time.time() - start
yhat = lr_best.predict(X_test)

from sklearn.linear_model import LogisticRegression as SKLogisticRegression
start = time.time()
lr_sk = SKLogisticRegression(C=1.0, penalty='l2', max_iter=1000, solver='liblinear')
lr_sk.fit(X_train, y_train)
sk_time = time.time() - start
yhat_sk = lr_sk.predict(X_test)

comparison = pd.DataFrame([
    {'model': 'Custom (Steepest Ascent)', 'accuracy': accuracy_score(y_test, yhat), 'training_time': custom_time},
    {'model': 'Sklearn (liblinear)', 'accuracy': accuracy_score(y_test, yhat_sk), 'training_time': sk_time}
])
print(comparison)

```

```python
# plot accuracy
fig = go.Figure(go.Bar(x=comparison['model'], y=comparison['accuracy']))
fig.update_layout(title='Accuracy: Custom vs Sklearn', yaxis_title='Accuracy', yaxis=dict(range=[0.8, 1.0]))
fig.show()

# plot training time
fig2 = go.Figure(go.Bar(x=comparison['model'], y=comparison['training_time']))
fig2.update_layout(title='Training Time: Custom vs Sklearn', yaxis_title='Seconds')
fig2.show()
```

Sklearn is generally more accurate and notably faster than our custom logistic regression. Initially, with sklearn's solver at lbfgs the accuracy was marginally higher at around .97 with a significantly increased training time of 5 seconds. This solver was unable to converge with the amount of iterations given, likely because our dataset is unscaled. Liblinear is better suited for unscaled data so we changed the solver to fix the issue.

Liblinear's tighter performance is likely due to it's sophisticated coordinate descent, where it's convergence is handled by optimizing one weight at a time. Steepest ascent updates all weights at once, leading to a less efficient convergence which likely factors into the difference in time and accuracy seen here.
